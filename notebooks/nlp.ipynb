```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem.porter import PorterStemmer
import pandas as pd
```

```python
stemmer = PorterStemmer()
 
def stem_words(text):
    # Split the input sentence into words
    word_tokens = text.split()
    # Stem each word individually
    stems = [stemmer.stem(word) for word in word_tokens]
    # Join the stemmed words back into a sentence
    stemmed_sentence = ' '.join(stems)
    return stemmed_sentence
def to_lower(value):
    # Process the value and return the result
    # For example, convert the value to lowercase
    return value.lower()
 
text = 'text preprocessing techniques for natural language processing by Aysel Aydin'
stem_words(text)
```

```python
import nltk
from nltk.corpus import stopwords

# Download the list of stopwords if not already downloaded
nltk.download('stopwords')

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = text.split()
    return ' '.join([word for word in words if word.lower() not in stop_words])
combined['description'] = combined['description'].apply(remove_stopwords)
```

```python
combined['stem']=combined['description'].apply(stem_words)
```

```python
column_list = combined['stem'].tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=1000)

X = vectorizer.fit_transform(column_list)

feature_names = vectorizer.get_feature_names_out()

X_array = X.toarray()

print("Unique Word List: \n", feature_names)
print("Bag of Words Matrix: \n", X_array)
```

```python
import pandas as pd

df = pd.DataFrame(data=X_array, columns=feature_names,index=column_list)
```

```python
X=df
Y=combined['scam']
```

```python
Y
```

```python
#train test split
from sklearn.model_selection import train_test_split
from sklearn import model_selection, naive_bayes, svm
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
print('BOW_cv_train:',X_train.shape)
print('BOW_cv_test:',X_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
#training the model with CountVectorizer features
mnb=MultinomialNB()
#fitting the nb for bag of words
mnb_bow=mnb.fit(X_train,Y_train)
print(mnb_bow)
```

```python
mnb_bow_predict=mnb.predict(X_test)
```

```python
mnb_bow_score=accuracy_score(Y_test,mnb_bow_predict)
print("mnb_bow_score :",mnb_bow_score)
```

```python
mnb_bow_report = classification_report(Y_test,mnb_bow_predict,target_names = ['0','1'])
print(mnb_bow_report)
cm_cv = confusion_matrix(Y_test,mnb_bow_predict)
cm_cv
```

```python
# Get the feature log probabilities from the trained Naive Bayes model
feature_log_probs = mnb.feature_log_prob_

# Print the shape of the feature log probabilities array
print("Shape of feature log probabilities:", feature_log_probs.shape)

# Get feature names from the vectorizer
feature_names = vectorizer.get_feature_names_out()

# Create a dictionary mapping feature names to log probabilities
feature_log_probs_dict = dict(zip(feature_names, feature_log_probs[1]))  # Assuming binary classification, index 1 for positive class

# Sort feature log probabilities by value
sorted_features = sorted(feature_log_probs_dict.items(), key=lambda x: x[1], reverse=True)

# Print the top N features with highest log probabilities
top_n = 10
for feature, log_prob in sorted_features[:top_n]:
    print(f"Feature: {feature}, Log Probability: {log_prob}")
```

```python
SVM = svm.SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
SVM.fit(X_train,Y_train)
```

```python
predictions_SVM = SVM.predict(X_test)
```

```python
print("SVM Accuracy Score -> ",accuracy_score(predictions_SVM, Y_test)*100)
```

```python
# Assuming you have trained an SVM model named 'svm_model'

# Get the coefficients assigned to each feature
coefficients = SVM.coef_

# Assuming you have a vectorizer for feature extraction
feature_names = vectorizer.get_feature_names_out()

# Create a dictionary to map feature names to coefficients
feature_coefficients = dict(zip(feature_names, coefficients[0]))

# Sort the feature coefficients by their absolute values
sorted_features = sorted(feature_coefficients.items(), key=lambda x: abs(x[1]), reverse=True)

# Print or analyze the top features
for feature, coefficient in sorted_features[:10]:
    print(f"Feature: {feature}, Coefficient: {coefficient}")
```
