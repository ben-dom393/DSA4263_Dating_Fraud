```python
import pandas as pd 
import numpy as np 
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
results=pd.read_csv("../models/combined_results.csv")
```

```python
results
```

# Using the combined_prob


Taking threshold as 0.5

```python
# 1. Create a column where the real label is 1 if either of the true columns says 1
results['real_label'] = ((results['true_profile_scam'] == 1) | (results['true_face_fake'] == 1)).astype(int)

# 2. Create a column where combined_prob more than 0.5 is 1, else 0
results['combined_label'] = (results['combined_prob'] > 0.5).astype(int)

# 3. Create a column where weighted_prob more than 0.5 is 1, else 0
results['weighted_label'] = (results['weighted_prob'] > 0.5).astype(int)
```

```python
results
```

```python


# Calculate metrics
accuracy = accuracy_score(results['real_label'], results['combined_label'])
precision = precision_score(results['real_label'], results['combined_label'])
recall = recall_score(results['real_label'], results['combined_label'])
conf_matrix = confusion_matrix(results['real_label'], results['combined_label'])
false_positive_rate = conf_matrix[0, 1] / (conf_matrix[0, 0] + conf_matrix[0, 1])
false_negative_rate = conf_matrix[1, 0] / (conf_matrix[1, 0] + conf_matrix[1, 1])

# Create DataFrame
metrics_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'False Positive Rate', 'False Negative Rate'],
    'Value': [accuracy, precision, recall, false_positive_rate, false_negative_rate]
})

# Print DataFrame
print(metrics_df)
```

```python
# Calculate metrics
accuracy = accuracy_score(results['real_label'], results['weighted_label'])
precision = precision_score(results['real_label'], results['weighted_label'])
recall = recall_score(results['real_label'], results['weighted_label'])
conf_matrix = confusion_matrix(results['real_label'], results['weighted_label'])
false_positive_rate = conf_matrix[0, 1] / (conf_matrix[0, 0] + conf_matrix[0, 1])
false_negative_rate = conf_matrix[1, 0] / (conf_matrix[1, 0] + conf_matrix[1, 1])

# Create DataFrame
metrics_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Precision', 'Recall', 'False Positive Rate', 'False Negative Rate'],
    'Value': [accuracy, precision, recall, false_positive_rate, false_negative_rate]
})

# Print DataFrame
print(metrics_df)
```
